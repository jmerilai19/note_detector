```python
import numpy as np
import matplotlib.pyplot as plt
import librosa
import librosa.display

def plot_audio_and_frequency_spectrum(audio_file):
    y, sr = librosa.load(audio_file, sr=None)
    
    print(f"Sample rate: {sr}, Audio duration: {len(y) / sr} seconds")

    plt.figure(figsize=(14, 8))

    plt.subplot(2, 1, 1)
    librosa.display.waveshow(y, sr=sr)
    plt.title('Waveform of Audio File')
    plt.xlabel('Time (s)')
    plt.ylabel('Amplitude')
    plt.grid()

    N = len(y)
    yf = np.fft.fft(y)
    xf = np.fft.fftfreq(N, 1 / sr)

    idx = np.arange(N // 2)
    xf = xf[idx]
    yf = np.abs(yf[idx])

    plt.subplot(2, 1, 2)
    plt.plot(xf, yf, label='Frequency Spectrum')

    c_freq = 261.63
    plt.axvline(c_freq, color='r', linestyle='--', label='C note frequency (261.63 Hz)')

    plt.xlim(0, 5000)

    plt.title('Frequency Spectrum (Limited to 5000 Hz)')
    plt.xlabel('Frequency (Hz)')
    plt.ylabel('Amplitude')
    plt.legend()
    plt.grid()

    plt.tight_layout()
    plt.show()

plot_audio_and_frequency_spectrum(audio_file)
```
